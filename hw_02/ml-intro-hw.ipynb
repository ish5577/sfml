#### Задание 3. Фунция предобработки

Напишите функцию, которая бы

* Удаляло идентификатор `AGREEMENT_RK`
* Избавлялась от проблем с '.' и ',' в стобцах PERSONAL_INCOME, CREDIT, FST_PAYMENT, LOAN_AVG_DLQ_AMT, LOAN_MAX_DLQ_AMT
* Что-то делала с пропусками
* Кодировала категориальные признаки

В результате, ваш датафрейм должен содержать только числа и не содержать пропусков!

```python
cat_to_remove = ['ORG_TP_STATE',
 'ORG_TP_FCAPITAL',
 'JOB_DIR',
 'FAMILY_INCOME',
 'PERSONAL_INCOME',
 'REG_ADDRESS_PROVINCE',
 'FACT_ADDRESS_PROVINCE',
 'POSTAL_ADDRESS_PROVINCE',
 'TP_PROVINCE',
 'REGION_NM',
 'CREDIT',
 'FST_PAYMENT',
 'LOAN_AVG_DLQ_AMT',
 'LOAN_MAX_DLQ_AMT',
 'sample']
for i in cat_to_remove:
    cat_names.remove(i)
#определяем Features, которые должны будут войти в DF и быть разбиты по методу get_dumnies
cat_names
```

```python

def preproc_data(df_input):
    df_output = df_input.copy()
    ## Your Code Here
    #убираем AGREEMENT_RK из DataFrame
    df_output.drop(['AGREEMENT_RK'], axis=1)
    #приводим в числовой формат все объекты Features, которые могут быть числовыми
    df_output['PERSONAL_INCOME'].map(lambda x: x.replace(',', '.')).astype('float')
    df_output['CREDIT'].map(lambda x: x.replace(',', '.')).astype('float')
    df_output['FST_PAYMENT'].map(lambda x: x.replace(',', '.')).astype('float')
    df_output['LOAN_AVG_DLQ_AMT'].map(lambda x: x.replace(',', '.')).astype('float')
    df_output['LOAN_MAX_DLQ_AMT'].map(lambda x: x.replace(',', '.')).astype('float')
    #работаем с категориальными данными по методу get_dumnies
    df_output = pd.get_dummies(df_output, columns = cat_names)
    #пойдем простым путем (надо же сдать домашнее задание): убираем все неполные Features
    length = len(df_input)
    for i in df_output.columns:
        if len(df_output[i]) != length:
            df_output = df_output.drop(i, axis = 1)
         
    return df_output
```

```python
df_preproc = df.pipe(preproc_data)

df_train_preproc = df_preproc.query('sample == "train"').drop(['sample'], axis=1)
df_test_preproc = df_preproc.query('sample == "test"').drop(['sample'], axis=1)
```
